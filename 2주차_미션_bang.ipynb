Q3.

```python
# 서울 소재 병원 중 상권업종소분류명에 "피부", "성형"이 들어간 데이터
df_seoul = df[df["시도명"] == "서울특별시"]
df_seoul_hospital = df_seoul[df_seoul["상권업종중분류명"] == "병원"]
df_seoul_dermatology_index = df_seoul_hospital["상호명"].str.contains("피부|성형")
df_seoul_dermatology = df_seoul_hospital[df_seoul_dermatology_index]
df_seoul_dermatology
```

```python
# 피부과 성형외과 수 / 전체병원 비율 구하기
df_seoul_hospital.loc[df_seoul_dermatology_index, "시군구명"].value_counts(normalize=True).round(2)
```

Q4.

```python
# 노인/치매병원 위치하는 곳 지도에 나타내기
df_senior_hospital = df.loc[df.상권업종소분류명 == "노인/치매병원",["상호명","경도","위도"]]

import folium

map_ans = folium.Map(location=[38,128], zoom_start=6)

for i in range(df_senior_hospital.shape[0]):
    folium.Marker([df_senior_hospital.iloc[i,2],df_senior_hospital.iloc[i,1]],
                  popup = df_senior_hospital.iloc[i,0]
                  ).add_to(map_ans)
    
map_ans
```
